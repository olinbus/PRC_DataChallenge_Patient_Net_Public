# FAA Aircraft Characteristics Data
https://www.faa.gov/airports/engineering/aircraft_char_database/data

```python
df_faa = pd.read_excel(os.path.join(strLocPath,"FAA-Aircraft-Char-DB-AC-150-5300-13B-App-2023-09-07.xlsx"))
```

```python
df_faa
```

```python
df_faa.columns
```

```python
df_faa_sel = df_faa[['ICAO_Code', 'MTOW_lb', 'MALW_lb', 'Wingspan_ft_without_winglets_sharklets', 'Wingspan_ft_with_winglets_sharklets', 'Length_ft']]
df_faa_sel
```

```python
df_faa_sel = df_faa_sel.rename(columns={"ICAO_Code": "aircraft_type"})
df_faa_sel['MTOW'] = df_faa_sel.pop('MTOW_lb') / 2.20462262185
df_faa_sel['MLW'] = df_faa_sel.pop('MALW_lb') / 2.20462262185

df_faa_sel['ac_L'] = df_faa_sel.pop('Length_ft') * 0.3048
df_faa_sel['ac_B'] = df_faa_sel.pop('Wingspan_ft_without_winglets_sharklets')
df_faa_sel['ac_B'] = df_faa_sel['ac_B'].fillna(df_faa_sel.pop('Wingspan_ft_with_winglets_sharklets'))
df_faa_sel['ac_B'] = df_faa_sel['ac_B'] * 0.3048


#df_faa_sel = df_faa_sel.drop(columns=['MTOW_lb', 'MALW_lb', 'Wingspan_ft_without_winglets_sharklets', 'Wingspan_ft_with_winglets_sharklets'])
df_faa_sel
```

```python
df_faa_sel.to_csv(os.path.join(strLocPath,'FAA-Aircraft-Char-DB-AC-150-5300-13B-App-2023-09-07_simple.csv'), index=False) 
```
